```python
# Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Crear la base de datos ficticia
data = {
    'Fecha de Venta': pd.date_range(start='2023-01-01', periods=100, freq='D'),
    'Región': ['Norte', 'Sur', 'Este', 'Oeste'] * 25,
    'Categoría de Producto': ['Electrónica', 'Ropa', 'Hogar', 'Alimentos'] * 25,
    'Cantidad Vendida': [5, 10, 15, 20, 25] * 20,
    'Precio por Unidad': [100, 50, 150, 200, 75] * 20
}

# Convertir a DataFrame
df = pd.DataFrame(data)

# Calcular los ingresos totales por venta
df['Ingresos Totales'] = df['Cantidad Vendida'] * df['Precio por Unidad']

# Mostrar los primeros registros
print("Datos de ventas:")
df.head()

# Análisis de las ventas por región
ventas_por_region = df.groupby('Región')['Ingresos Totales'].sum()
print("\nIngresos Totales por Región:")
print(ventas_por_region)

# Visualizar las ventas por región
plt.figure(figsize=(10, 6))
sns.barplot(x=ventas_por_region.index, y=ventas_por_region.values)
plt.title('Ingresos Totales por Región')
plt.xlabel('Región')
plt.ylabel('Ingresos Totales')
plt.show()

# Análisis de las ventas por categoría de producto
ventas_por_categoria = df.groupby('Categoría de Producto')['Ingresos Totales'].sum()
print("\nIngresos Totales por Categoría de Producto:")
print(ventas_por_categoria)

# Visualizar las ventas por categoría de producto
plt.figure(figsize=(10, 6))
sns.barplot(x=ventas_por_categoria.index, y=ventas_por_categoria.values)
plt.title('Ingresos Totales por Categoría de Producto')
plt.xlabel('Categoría de Producto')
plt.ylabel('Ingresos Totales')
plt.show()

# Análisis de tendencias de ventas a lo largo del tiempo
ventas_por_fecha = df.groupby('Fecha de Venta')['Ingresos Totales'].sum()

# Visualizar las ventas a lo largo del tiempo
plt.figure(figsize=(14, 8))
plt.plot(ventas_por_fecha.index, ventas_por_fecha.values, marker='o')
plt.title('Ingresos Totales a lo Largo del Tiempo')
plt.xlabel('Fecha de Venta')
plt.ylabel('Ingresos Totales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
