```python
# Importar las bibliotecas necesarias
import pandas as pd
import numpy as np

# Crear una base de datos de ejemplo
data = {
    'Nombre': ['Juan Pérez', 'Ana López', 'Carlos Martínez', 'María Gómez', 'Luis Rodríguez', 'Nombre Desconocido', None],
    'Edad': [25, 30, 45, None, 55, 30, 29],
    'Salario': [50000, 60000, 80000, 75000, None, 62000, 58000],
    'Departamento': ['Ventas', 'Marketing', None, 'IT', 'HR', 'Marketing', 'IT']
}

# Convertir a DataFrame
df = pd.DataFrame(data)

# Mostrar los primeros registros
print("Datos originales:")
df.head()
```

```python
# Eliminar filas con nombres faltantes
df = df.dropna(subset=['Nombre'])

# Rellenar valores nulos en la columna 'Edad' con la mediana
df['Edad'] = df['Edad'].fillna(df['Edad'].median())

# Rellenar valores nulos en la columna 'Salario' con el promedio
df['Salario'] = df['Salario'].fillna(df['Salario'].mean())

# Rellenar valores nulos en la columna 'Departamento' con 'Sin Asignar'
df['Departamento'] = df['Departamento'].fillna('Sin Asignar')

# Mostrar los datos limpios
print("Datos después de la limpieza:")
df.head()
```

```python
# Exportar los datos limpios a un archivo CSV
df.to_csv('empleados_limpios.csv', index=False)
```
